#### Setting up functions for similar graphs across the notebook

```python
# Set global seaborn style
sns.set_theme(style="darkgrid", palette="pastel", context="notebook")

def draw_barplot(data, x, y, xlabel, ylabel, title, hue=None, palette=None):
    """
    Function to draw a barplot
    
    Parameters:
    - data: DataFrame or Series to plot.
    - x: str, column name for x-axis.
    - y: str, column name for y-axis.
    - xlabel: str, label for x-axis.
    - ylabel: str, label for y-axis.
    - title: str, title of the plot.
    - hue: str, column name for grouping bars.
    - palette: list or str, optional. The color palette for the bars.
    """
    plt.figure(figsize=(10, 6))
    sns.barplot(x=x, y=y, hue=hue, data=data, palette=palette, alpha=0.7)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.legend(title=hue if hue else "")
    plt.show()
```

```python
def draw_pie_chart(sizes, labels, title, colors=None, explode=None):
    """
    Function to draw a pie chart

    Parameters:
    - sizes: list of numerical values representing the sizes of each slice.
    - labels: list of str, the labels for each slice.
    - title: str, title of the pie chart.
    - colors: list of str, optional. The colors for each slice.
    - explode: list of float, optional. The fraction of the radius to offset each slice.
    """
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.title(title)
    plt.axis('equal')
    plt.show()
```

```python
def draw_stacked_barplot(data, xlabel, ylabel, title, colors, legend_labels):
    """
    Function to draw a stacked bar plot

    Parameters:
    - data: DataFrame to plot, indexed appropriately.
    - xlabel: str, label for x-axis.
    - ylabel: str, label for y-axis.
    - title: str, title of the plot.
    - colors: list of str, colors for each stack.
    - legend_labels: list of str, labels for the legend.
    """
    data.plot(kind='bar', stacked=True, color=None, figsize=(10, 6))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(title='Heart Attack', labels=legend_labels)
    plt.xticks(rotation=0)
    plt.show()
```

```python
def draw_countplot(data, x, xlabel, ylabel, title, hue=None, palette=None):
    """
    Function to draw a count plot
    
    Parameters:
    - data: DataFrame to plot.
    - x: str, column name for x-axis.
    - xlabel: str, label for x-axis.
    - ylabel: str, label for y-axis.
    - title: str, title of the plot.
    - hue: str, column name for grouping bars (optional).
    - palette: list or str, optional. The color palette for the bars.
    """
    plt.figure(figsize=(8, 6))
    ax = sns.countplot(data=data, x=x, hue=hue, palette=palette)
    
    # Annotate each bar with count and percentage
    total = len(data)
    for p in ax.patches:
        count = int(p.get_height())
        percentage = (count / total) * 100
        ax.annotate(f'{count} ({percentage:.2f}%)', (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom', fontsize=12)
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=0)
    plt.show()
```

```python
def draw_grouped_barplot(data, xlabel, ylabel, title, colors=None, legend_title=None):
    """
    Function to draw a grouped bar plot

    Parameters:
    - data: DataFrame to plot, containing values for each group.
    - xlabel: str, label for x-axis.
    - ylabel: str, label for y-axis.
    - title: str, title of the plot.
    - colors: list of str, colors for each group.
    - legend_title: str, title for the legend (optional).
    """
    ax = data.plot(kind='bar', figsize=(10, 6), color=colors)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.title(title, fontsize=16)
    plt.xticks(rotation=0, fontsize=12)
    plt.yticks(fontsize=12)
    if legend_title:
        plt.legend(title=legend_title)
    plt.grid(True)
    plt.show()
```

***ANALYSIS*** : Exercise is one of the major known prescipitants of chest pain. We wanted to understand if this feature was prominent in the patients having heart attacks from our dataset. Indeed, almost 80% of patients who had a heart attack, did reported angina brought up by exercise.

Exercise-induced or provoked angina is a symptom of ischaemic heart disease (IHD) and is associated with an increased risk of heart attacks (myocardial infarction). Several studies have explored this relationship, highlighting the connection between exercise-provoked angina, underlying ischemia, and cardiovascular events.

This is one of the basis of using "treadill test" or "cardiac stress test". Patients with exercise-induced angina who develop ischemia during cardiac stress testing are at a significantly higher risk of heart attacks. The inability of coronary arteries to supply sufficient blood during exercise leads to myocardial ischemia, which, if persistent, can result in a heart attack (1-2).

Medical literature supports the notion that patients with exercise-induced angina show a higher risk for major adverse cardiac events (MACE), including myocardial infarction and death, compared to those without such symptoms during stress testing (3). In fact, angina, especially when provoked by exercise, is a key indicator of coronary artery blockage. Such blockages can rupture, leading to a thrombotic event, which results in a heart attack (4).

Later in this notebook, in our diagnostic analysis, we will look at how strong or not is the relationship between exercise-provoked angina and an elevated risk of heart attacks, which in turn could be a key feature for early detection of ischaemic heart disease.

Another interesting observation in patients who did not have a heart attack:  40% did experience pain but it was non anginal. 30% were asymptomatic.

Atypical angina refers to chest pain or discomfort that does not fit the classic description of angina or the characteristic pain caused by myocardial ischemia. The relationship between atypical angina and heart attack (myocardial infarction) risk has been explored in various studies, with results showing that although atypical angina presents a lower immediate risk compared to typical angina, it still represents a significant marker for underlying coronary artery disease (CAD) and future cardiovascular events. Clinical practice and litearature supports the notion that atypical chest pain symptoms still had a considerable risk of developing coronary artery disease and myocardial infarction, and therefore should not be disregarded, especially in patients with other cardiovascular risk factors (1).

Research had pointed out that patients with atypical angina are at risk of underdiagnosis or delayed diagnosis, which may contribute to worse outcomes when myocardial infarction occurs. Although the chest pain is less predictable, patients presenting with atypical symptoms are at an increased risk of CAD and future heart attacks (2). Furthermore, patients with atypical angina, while having a lower short-term risk of myocardial infarction compared to those with typical angina, still require thorough diagnostic workups due to the potential for underlying ischemia and CAD. Unaddressed, these patients remain at risk for heart attacks, especially if their symptoms progress (3).

Asymptomatic patients who did not have chest pain or angina despite having a heart attack are described in the literature as having a "silent" heart attack. Various medical studies have explored reasons why some patients, particularly certain groups, may not experience chest pain during a heart attack. A well known group of patients in this category are diabetic patients. Diabetes is a well established cause of silent heart attacks. Patients with diabetes often develop neuropathy, which can impair the nerves responsible for detecting pain in the chest. As a result, these patients may not feel the pain associated with ischemia or infarction (4). Other well known group who more likely have silent heart attacks are older adults. With age, the sensitivity to pain diminishes, and older individuals might not recognize or feel chest pain as younger people do (5). Gender differences had also been stablished. Women are more likely to experience non pain related symptoms, such as fatigue, nausea, or shortness of breath, rather than chest pain. Hormonal differences, especially in younger women, are thought to play a role (6). Lastly, Other factors which could explain silent heart attacks are lack of awareness of chest pain as a symptom (Patients may attribute symptoms like indigestion, fatigue, or mild discomfort to other less serious conditions, thereby not recognizing the onset of a heart attack) and differences in pain perception (7).

These factors indicate that many patients, especially those with diabetes, older adults, and women, may not experience the classic symptom of chest pain during a heart attack, which makes early detection and prevention strategies challenging to rely on this symptom, particularly given our dataset was imbalance in terms of gender in patients who had heart attacks (90% were males) and the average age of having heart attacks was 57-58 years old.

How this feature  help the ML algorithm to predict heart attack risk will need to be seen with care in our dashboard development given the interesting type of pain patterns patients in our dataset had.

### Interpretation of Feature Distribution

- **Resting Heart Rate**, **Age**, **Serum Cholesterol**, and **Max Heart Rate**:
  - These features are **biological measurements** and exhibit distributions that are **approximately normal**.
  - For these features, **Standardization** is the most appropriate scaling method. By standardizing, we ensure that these features have a **mean of 0** and a **standard deviation of 1**, which helps make them comparable for clustering.

- **Binary Features**:
  - Features like `high_fasting_blood_sugar`, `exercise_induced_angina`, and `family_history_cad` are **binary** and are naturally represented as **0 or 1**.
  - These features will be **passed through** as-is, without applying additional scaling, as their current representation is suitable for clustering.

- **Categorical Features**:
  - Categorical features like `gender`, `chest_pain_type`, and `resting_ecg_results` will be transformed using **one-hot encoding**.
  - This transformation converts each category into a **binary feature** with values **0 or 1**. This helps to avoid introducing any unintended ordinal relationships between the categories.

By applying **Standardization** to continuous features, retaining **0/1** for binary features, and using **one-hot encoding** for categorical variables, we can ensure that each feature type is appropriately prepared for clustering, enabling balanced distance calculations and meaningful clusters.


```python
# Define categorical and numerical features
binary_features = ['high_fasting_blood_sugar', 'exercise_induced_angina', 'family_history_cad']
categorical_features = ['gender', 'chest_pain_type', 'resting_ecg_results']
numerical_features = ['age', 'serum_cholesterol', 'max_heart_rate', 'st_depression', 
                      'has_hypertension', 'cigarettes_per_day', 'years_smoking', 'resting_heart_rate']

# Apply one-hot encoding only to non-binary categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features),
        ('bin', 'passthrough', binary_features)  # Keep binary features as-is
    ])

# Transform the data
data_preprocessed = preprocessor.fit_transform(data)

# Extract column names for the one-hot encoded features
one_hot_encoder = preprocessor.named_transformers_['cat']
encoded_columns = one_hot_encoder.get_feature_names_out(categorical_features)

# Combine numerical, binary, and encoded categorical feature names
all_columns = numerical_features + binary_features + list(encoded_columns)

# Convert to DataFrame for easy visualization (optional)
data_preprocessed_df = pd.DataFrame(data_preprocessed, columns=all_columns)
```

Firstly, the data needs to be prepared for clustering.

```python
# Remove 'Cluster' column if it exists to avoid including labels in clustering
data_for_clustering = data_preprocessed_df.drop(columns=['Cluster'], errors='ignore')

# Calculate silhouette scores for different values of k
silhouette_scores = []
K = range(2, 11)  # Start from k=2 since silhouette score is not defined for k=1

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_for_clustering)
    score = silhouette_score(data_for_clustering, kmeans.labels_)
    silhouette_scores.append(score)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(K, silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(K)
plt.grid()
plt.show()
```

```python
optimal_k = 2

# Perform KMeans clustering with k=2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(data_for_clustering)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data_for_clustering)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=reduced_data[:, 0], y=reduced_data[:, 1], hue=clusters, palette='viridis', s=60, alpha=0.7)
plt.title('2D Visualization of Clusters (k=2)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
# Add heart disease diagnosis to the reduced data and transform it to a df
reduced_df = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])
reduced_df['Cluster'] = clusters
reduced_df['Heart_Disease'] = data['heart_disease_diagnosis']

cluster_mapping = {}

for cluster in np.unique(clusters):
    # Get the true labels for samples in this cluster
    true_labels = reduced_df[reduced_df['Cluster'] == cluster]['Heart_Disease']
    
    # Find the unique values and their counts
    values, counts = np.unique(true_labels, return_counts=True)
    
    # Get the value with the highest count (majority class)
    majority_class = values[np.argmax(counts)]
    
    # Map the cluster label to the majority class
    cluster_mapping[cluster] = majority_class

# Relabel clusters based on majority class mapping
reduced_df['Predicted_Target'] = reduced_df['Cluster'].map(cluster_mapping)

# Calculate the percentage of correctly assigned labels
correct_predictions = reduced_df['Predicted_Target'] == reduced_df['Heart_Disease']
accuracy = correct_predictions.mean() * 100

# Print the result
print(f"Percentage of Correctly Clustered Patients: {accuracy:.2f}%")

# Create a scatter plot with clustering results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=reduced_df, palette='viridis', s=60, alpha=0.7)
plt.title('Clusters from KMeans (k=2)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')

# Create a scatter plot with original target
plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='Heart_Disease', data=reduced_df, palette='coolwarm', s=60, alpha=0.7)
plt.title('Original Heart Disease Diagnosis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Heart Disease')

plt.tight_layout()
plt.show()
```

### Interpretation of KMeans Clustering

The 2D visualization using PCA does not reveal clearly distinguishable clusters, indicating that the data may not form well-separated groups in a reduced dimensional space. The silhouette score, which peaked at around 0.13 for two clusters, is quite lowsuggesting that the clustering quality is not ideal. This implies that the dataset may not naturally exhibit distinct, well-defined clusters.

Despite these limitations, a comparison between the predicted cluster labels and the actual heart disease diagnosis labels showed an accuracy of 71.2%, which indicates that the clustering captured some meaningful relationships in the data. However, given the low silhouette score, this clustering result might not be reliable.

Therefore, another approach should be used to try to understand the data better.

### Hierachical Clustering

```python
# Use the preprocessed data without cluster labels
data_for_clustering = data_preprocessed_df.drop(columns=['Cluster'], errors='ignore')

# Create Linkage Matrix and Plot Dendrogram
linkage_matrix = linkage(data_for_clustering, method='ward')

# Plotting the dendrogram
plt.figure(figsize=(15, 8))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

### Evaluation of Dendrogram
The dendrogram displays too many samples with numerous endpoints, making it difficult to analyze the clusters in detail. However, a visual inspection of the graph suggests the presence of 2 to 3 distinct main clusters. These clusters should be further identified, and additional analysis should be performed to evaluate their characteristics and potential relationships with the original target labels

```python
# Only use the top two clusters
cluster_labels = fcluster(linkage_matrix, t=2, criterion='maxclust')

# Add the cluster labels to the original preprocessed DataFrame
data_preprocessed_df['Cluster'] = cluster_labels

# Add the original target column back to the preprocessed DataFrame for comparison
data_preprocessed_df['Original_Target'] = data['heart_disease_diagnosis']

# Calculate how well the clusters match
cluster_summary = data_preprocessed_df.groupby(['Cluster', 'Original_Target']).size().unstack(fill_value=0)
print("\nCluster vs Original Target Summary:")
print(cluster_summary)

# Step 1: Determine the majority class for each cluster
cluster_mapping = {}
for cluster in np.unique(cluster_labels):
    true_labels = data_preprocessed_df[data_preprocessed_df['Cluster'] == cluster]['Original_Target']
    values, counts = np.unique(true_labels, return_counts=True)
    majority_class = values[np.argmax(counts)]
    cluster_mapping[cluster] = majority_class

# Step 2: Assign predicted target based on the majority class for each cluster
data_preprocessed_df['Predicted_Target'] = data_preprocessed_df['Cluster'].map(cluster_mapping)

# Step 3: Calculate accuracy
correct_predictions = data_preprocessed_df['Predicted_Target'] == data_preprocessed_df['Original_Target']
accuracy = correct_predictions.mean() * 100
print(f"Percentage of Correctly Clustered Patients: {accuracy:.2f}%")

# Step 4: Plot comparison between clusters and target values
plt.figure(figsize=(10, 6))
sns.countplot(data=data_preprocessed_df, x='Cluster', hue='Original_Target', palette='viridis')
plt.title('Comparison of Clusters and Original Target (Heart Disease Diagnosis)')
plt.xlabel('Cluster')
plt.ylabel('Number of Patients')
plt.legend(title='Heart Disease Diagnosis')
plt.show()
```

The hierarchical clustering approach achieved a slightly lower accuracy compared to the KMeans method.
